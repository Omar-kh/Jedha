```
class ToDo:
  def __init__(self):
    self.task_list = []
  
  def add_task(self, task):
    self.task_list.append(task)
  def remove_task(self, task):
    self.task_list.remove(task)
  
  def show_tasks(self):
    print("\nThese are your tasks with their statuses\n")
    for task in self.task_list:
      print("\n{} : {}\n".format(task.task_title, task.task_status))
    
class Task:
  def __init__(self, task_title, status = 'Initialized'):
    self.task_title = task_title
    self.task_status = status
    
  def update_task_title(self, new_task_title):
    self.task_title = new_task_title
  def update_task_status(self, new_status):
    self.task_status = new_task_status
    
my_todo = ToDo()
first_task = Task('go home')
second_task = Task('drink water', 'Done')
third_task = Task('sleep', 'Impatiently waiting for it')

my_todo.add_task(first_task)
my_todo.add_task(second_task)
my_todo.add_task(third_task)

#print("Task title is {}, task status is {}".format(my_todo.task_list[0].task_title, my_todo.task_list[0].task_status))

#my_todo.task_list[0].update_task_title('not go home')

#print("Task title is {}, task status is {}".format(my_todo.task_list[0].task_title, my_todo.task_list[0].task_status))

my_todo.show_tasks()

my_todo.remove_task(second_task)

my_todo.show_tasks()
```
