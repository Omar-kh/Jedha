```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import sklearn
X_norm = sklearn.preprocessing.normalize(df.iloc[:,:-2], axis = 0) 
X = df.iloc[:,:-1]
```

```python
X
```

```python
Y = df.price
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y)
X_train_norm, X_test_norm, y_train_norm, y_test_norm = train_test_split(X, Y)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X, Y)
```

```python
reg.coef_
```

```python
coeffs = pd.DataFrame(data = reg.coef_, columns = ['Coefficients'], index = data.feature_names)
```

```python
coeffs.sort_values("Coefficients", ascending = False)
```

```python
from sklearn.feature_selection import f_regression
freg = f_regression(X, Y)
```

```python
freg_np = np.array(freg)
```

```python
freg_np.transpose()
```

```python
fcoeff = pd.DataFrame(data = freg_np.transpose(), columns = ["f-score", "p-value"], index = data.feature_names)
sns.barplot("f-score", fcoeff.index, data = fcoeff.sort_values("f-score", ascending = False), orient = "h")
```
