# Exercices

```python
house_pricing = pd.read_csv("./Data_for_exercises/house_pricing.csv")
house_pricing.info()
```

```python
house_pricing.columns
```

```python
house_pricing.head()
```

```python
house_pricing_filtered = house_pricing[(house_pricing.LotArea < 20000) & (house_pricing.SalePrice < 500000) ]
# Same as: house_pricing.query("LotArea < 20000 & SalePrice < 500000")
lot_area = house_pricing_filtered.LotArea.fillna(house_pricing_filtered.LotArea.mean())
sales_prices = house_pricing_filtered.SalePrice.fillna(house_pricing_filtered.SalePrice.mean())
sns.scatterplot(lot_area, sales_prices)
```

```python
house_pricing_filtered2 = house_pricing[(house_pricing.LotFrontage < 200) & (house_pricing.LotArea < 100000)]
lot_area2 = house_pricing_filtered2.LotArea.fillna(house_pricing_filtered2.LotArea.mean())
lot_frontage = house_pricing_filtered2.LotFrontage.fillna(house_pricing_filtered2.LotFrontage.mean())
sns.scatterplot(lot_frontage, lot_area2)
```

```python
new_house_pricing = house_pricing.iloc[:,-15:]
corr = new_house_pricing.corr()
corr
```

```python
plt.figure(figsize= (20, 10))
sns.heatmap(corr, annot=True) #keep MiscVal, YrSold, 3SsnPorch
```

```python
sales_prediction = pd.read_csv("./Data_for_exercises/sales_predictions.csv")
sales_prediction.head()
```

```python
sales_prediction.info()
```

```python
sales_prediction.date = sales_prediction.date.str.split('.')
```

```python
sales_prediction.date = sales_prediction.date.apply(lambda x: datetime.datetime(int(x[2]), int(x[1]), int(x[0])))
```

```python
sales_prediction["date_values"] = sales_prediction.date.apply(lambda x: x.timestamp())
```

```python
sales_prediction
```

```python
sales_prediction.drop(["date_values"], 1)
```

```python
sns.relplot("date", "item_price", data= sales_prediction, kind="line")
```
