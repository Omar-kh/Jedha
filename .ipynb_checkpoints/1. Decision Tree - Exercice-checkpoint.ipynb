# Exercice Decision Tree Classifier 

0. Importez les librairies usuelles 

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

1. Importez les data et les afficher

````python
df = pd.read_csv("Data_for_exercises/bl.txt")
df.head()
````

2. Afficher un résumé des statistiques descriptives

````python
df.describe()
````

3. Afficher la shape du dataset

````python
df.shape
````

4. Séparer le dataset en deux array ```X``` et ```Y```

```X``` désignant les variables explicatives / features & ```Y``` la variable cible / target

````python
X = df.iloc[:,1:]
y = df.iloc[:,:1]
````

5. Spliter le dataset (70/30) avec un random_state=42

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42, stratify=y)
````

6. Instancier un 'DecisionTreeClassifier' avec un critère de décision GINI

````python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
clf = DecisionTreeClassifier(criterion="gini", max_depth=4).fit(X_train, y_train)
````

7. Affichez votre arbre de decision

````python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()

export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
````

````python
import graphviz
graphviz.Source(dot_data.getvalue())
````

8. Instancier un 'DecisionTreeClassifier' avec un critère de décision "d'information gain"

````python
clf2 = DecisionTreeClassifier(criterion="entropy").fit(X_train, y_train)
````

9. Afficher votre arbre

````python
dot_data2 = StringIO()

export_graphviz(clf2, out_file=dot_data2,  
                filled=True, rounded=True,
                special_characters=True)

graph2 = pydotplus.graph_from_dot_data(dot_data2.getvalue())  
Image(graph2.create_png())
````

10. Vérifier que l'individue X_new est bien classé "Left"

où X_new est l'observation [4, 4, 3, 3]

````python
X_new = np.array([4,4,3,3])
clf.predict(X_new.reshape(1, -1))
````

11. Effectuer une prediction du classifieur GINI sur l'ensemble de test

````python
clf.predict(X_test)
````

12. Effectuer une prediction du classifieur ENTROPY sur l'ensemble de test

````python
clf2.predict(X_test)
````

12. Effectuer une comparaison entre les classifieurs GINI & ENTROPY

On prendra pour temoin le dataset de test

````python
pred1 = pd.Series(clf.predict(X_test))
pred2 = pd.Series(clf2.predict(X_test))
````

````python
pred_compare = pd.concat([pred1, pred2], axis=1)
````

````python
pred_compare.columns = ["pred1", "pred2"]
pred_compare
````

````python
pred_compare["pred1"].value_counts()
````

````python
pred_compare["diff"] = pred_compare["pred1"] == pred_compare["pred2"]
pred_compare["diff"].value_counts()
````

````python
1 - (pred_compare["diff"].sum() / len(pred_compare))
````

13. Donner l'accuracy du classifieur GINI

````python
clf.score(X_test, y_test)
````

14. Donner l'accuracy du classifieur Entropy 

````python
clf2.score(X_test, y_test)
````

15. Afficher la matrice de confusion du classifieur GINI

````python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test, clf.predict(X_test)), annot=True)
````
