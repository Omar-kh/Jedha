Nous allons étudier les facteurs qui déterminent l'admission d'un élève à un master en université américaine. 

Parmi ces facteurs, on compte : 

1. GRE Scores ( out of 340 )

2. Notre variable est exprimée comme une probabilité. On voudrait que ce soit plutôt une classe. Remplacez donc toutes les probabilités supérieures à 75% par 1 et les autres par 0 

````python
adm["Chance of Admit "] = adm["Chance of Admit "].apply(lambda x: 1 if x > 0.75 else 0)
````

6. Séparez votre dataset en variables explicatives : ```X``` & variable cible : ```y``` 

````python
X = adm.iloc[:, :-1]
y = adm.iloc[:,-1:]
````

7. Séparez votre dataset en données d'entrainement et de test 

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
````

8. Tentez de faire un premier arbre de décision et évaluez votre score dans une matrice de confusion

````python
from sklearn.tree import DecisionTreeClassifier
tr = DecisionTreeClassifier(criterion='gini').fit(X_train, y_train)
````

````python
tr.score(X_test, y_test)
````

9. Visualisez votre arbre via graphviz 

````python
import graphviz
from sklearn.tree import export_graphviz
graphviz.Source(export_graphviz(tr))
````

10. Classez les features les plus importantes qui déterminent les admissions 

````python
from sklearn import tree
help(tree._tree.Tree)
````

````python
features_score = pd.Series(tr.feature_importances_)
features = pd.Series(X.columns)
feature_by_score = pd.concat([features, features_score], axis=1)
````

````python
feature_by_score.columns = ['Feature', 'Score']
````

````python
feature_by_score.sort_values(by="Score" ,ascending = False, inplace = True)
````

````python
sns.barplot('Score', 'Feature', data = feature_by_score, orient="h")
````

11. Essayons d'améliorer notre score. Faites une Random Forest avec 50 arbres de decision

````python
from sklearn.ensemble import RandomForestClassifier
rtr50 = RandomForestClassifier(n_estimators=50, criterion="gini", bootstrap=True).fit(X_train, y_train)
````

````python
rtr50.score(X_test, y_test)
````

````python
rtr20 = RandomForestClassifier(n_estimators=20, criterion="gini", bootstrap=True).fit(X_train, y_train)
````

````python
rtr20.score(X_test, y_test)
````

12. Tentez de refaire le modèle avec uniquement 20 arbres 

---> Il semblerait qu'il y ait un nombre optimum d'arbres.

````python
[1,*np.arange(1,11)*10]
````

13. Faites une Grid Search pour trouver le nombre optimum d'arbres 

````python
from sklearn.model_selection import GridSearchCV
rtr = RandomForestClassifier(criterion="gini", bootstrap=True, random_state=42)
params = {"n_estimators": [1,*np.arange(1,11)*10], "max_depth": np.arange(1,10)}
rtrgrid = GridSearchCV(rtr, param_grid=params, cv=5).fit(X_train, y_train)
````

````python
rtrgrid.best_estimator_.score(X_test, y_test)
````

14. Regardez les feature importance 

````python
rtrgrid.best_estimator_.feature_importances_
````

````python
rtrgrid.best_params_
````

````python
grid_features_score = pd.Series(rtrgrid.best_estimator_.feature_importances_)
grid_features = pd.Series(X.columns)
grid_feature_by_score = pd.concat([grid_features, grid_features_score], axis=1)
grid_feature_by_score.columns = ["Feature","Score"]
grid_feature_by_score.sort_values(by="Score", inplace = True, ascending=False)
````

````python
sns.barplot("Score", "Feature", data=grid_feature_by_score, orient="h")
````

````python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test, rtrgrid.best_estimator_.predict(X_test)), annot=True)
````
