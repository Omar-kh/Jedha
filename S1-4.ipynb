### House market exercise

```python
nb_sdb = pd.read_csv("./Data_for_exercises/nombre_de_sdb.csv", 
                  sep=',', 
                  names=["id", "nb_sdb"])
nb_rooms = pd.read_csv("./Data_for_exercises/nombre_de_chambres.csv", 
                  sep=',', 
                  names=["id", "nb_rooms"])
house_surface = pd.read_csv("./Data_for_exercises/superficie.csv", 
                  sep=',', 
                  names=["id", "Surface"])
house_price = pd.read_csv("./Data_for_exercises/house_price.csv")
house_price.set_index('id')
```

```python
dataset = pd.concat([house_surface.set_index('id'), nb_sdb.set_index('id'), nb_rooms.set_index('id'), house_price.set_index('id')], axis = 1, join = 'inner').reset_index()
dataset
```

```python
np.mean(dataset.Surface)
```

```python
dataset.groupby('nb_rooms').median()
```

```python
dataset.groupby('nb_rooms').mean()
```

```python
dataset['house_price'].mean()
```

```python
dataset.groupby("house_price").mean()
```

```python
dataset[dataset.loc[:,"nb_rooms"] == 1].mean()
```

```python
def set_house_size(surface):
    if surface > 25000:
        return "TrÃ¨s Grande"
    elif surface > 20000:
        return "Grande"
    elif surface > 15000:
        return "Moyenne"
    elif surface > 10000:
        return "Petite"
    elif surface <= 10000:
        return "TrÃ¨s petite"
    else:
        return "Unknown"
dataset["house_size"] = dataset.Surface.apply(lambda x: set_house_size(x))
dataset
```

```python
# Alternative
dataset["house_size"] = dataset.Surface.apply(lambda x: "TrÃ¨s Grande" if x > 25000
                                             else "Grande" if x > 20000
                                             else "Moyenne" if x > 15000
                                             else "Petite" if x > 10000
                                             else "TrÃ¨s petite")
```
